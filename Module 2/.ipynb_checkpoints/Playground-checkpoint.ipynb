```python
from scipy.stats import uniform
import numpy as np
import matplotlib.pyplot as plt
```

```python
# loc is the left edge of the distribution (a)
# scale is the length (b - a)
U = uniform(loc=10, scale=5)
```

```python
upoints = np.linspace(7, 18, 200)

ppoints = U.pdf(upoints)

plt.plot(upoints, ppoints)
```

```python
U.mean()
```

```python
U.var()
```

```python
U.std()
```

```python
U.rvs(size=3)
```

```python
plt.hist(U.rvs(size=1000000), density=True)
plt.plot(upoints, ppoints)
```
