### In this exercise, generate some basic charts using Pandas. 


```python
# import data from pandas
import pandas as pd
import os

#Make directory called plots
os.makedirs("plots", exist_ok=True)

#create the DataFrame utilizing Pandas 
data = pd.read_csv("https://media.geeksforgeeks.org/wp-content/uploads/nba.csv") 

#generate all info for this dataset
data.info()

#Remove the last line of NAN
data = data.iloc[:-1 , :]

#Ignore any warnings
import warnings
warnings.filterwarnings('ignore')
```

###  Part One: In this exercise, generate a basic histogram.


```python
#Please generate a basic histogram and save the histogram as a png file
# Hint: use pandas histogram function.

ax = data.plot.hist()
ax.figure.savefig('plots/demo-file.png')
```

###  Part Two: In this exercise, generate a basic scatter plot. 


```python
#Please generate a basic scatter plot utilizing Age as the x axis and Salary as the y axis
# Hint: use pandas scatter function to generate the chart.

scat = data.plot.scatter(x='Age', y='Salary')

print(scat)


```

###  Part Three: In this exercise, generate a basic lineplot. 

```python
# Please generate a basic line plot with Name as the x axis and Salary as the y axis 
# for the "Golden State Warriors" using the nba data set. 
# Note a new dataframe called teamdata has been created for you that will be used during the rest of this exercise



teamdata = data.filter(items=['Team', 'Name', 'Salary'])
teamdata = teamdata.loc[data.Team == 'Golden State Warriors', 'Name':'Salary']
teamdata.set_index('Name')


lines = teamdata.plot.line(x='Name', y='Salary')


# ANSWER CHECK
print(lines)
```

###  Part Four: In this exercise, generate a basic bar chart. 

```python
# Please generate a basic bar plot with Name as the x axis and Salary as the y axis of the teamdata dataframe.
 

bar = teamdata.plot.bar(x='Name', y='Salary')

print(bar)

```

```python
# Create a plot showing difference in most used platform amongst countries

# Could do a bar chart between the two?

# First, filter down the data to just Country and platform
data = data[['Country', 'Most_Used_Platform', ]]
data
```

```python
data = data[data['Most_Used_Platform'] == 'TikTok']
data
```

```python
data['Count'] = 1

pivot_sum = pd.pivot_table(data, values='Count', index='Country', aggfunc='sum')
pivot_sum
# type(pivot_sum)
# data.plot.bar(x='Gender', y='dummy')
# data.sum()
```

```python
pivot_sum.plot.bar(title='Frequency of TikTok being the most used platform')
```
