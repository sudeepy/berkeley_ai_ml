```python
import pandas as pd
```

```python
df = pd.read_csv('data/States_and_union_territories_of_India_1.csv')
df
```

```python
df = df.rename(columns={
    'Officiallanguages': 'official',
    'Additional officiallanguages': 'other',
    'Population(2011)': 'population',
})
df['population'] = pd.to_numeric(df['population'].str.replace(',', ''))
df
```

```python
telugu_states = df[(df['official'].str.contains('Telugu')) | (df['other'].str.contains('Telugu'))]
telugu_states
```
